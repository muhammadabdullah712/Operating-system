```python
class NaiveBayesClassifier:
    def _init_(self):
        self.prior = {}
        self.likelihood = {}
        self.classes = None

    def fit(self, X, y):
        self.classes = list(set(y))
        for cls in self.classes:
            X_cls = [X[i] for i in range(len(X)) if y[i] == cls]
            self.prior[cls] = len(X_cls) / len(X)
            self.likelihood[cls] = {}
            for col in range(len(X[0])):
                values = [X_cls[i][col] for i in range(len(X_cls))]
                unique_values = list(set(values))
                self.likelihood[cls][col] = {val: values.count(val) / len(values) for val in unique_values}

    def predict(self, X):
        predictions = []
        for x in X:
            posteriors = {}
            for cls in self.classes:
                posterior = self.prior[cls]
                for col in range(len(x)):
                    if x[col] in self.likelihood[cls][col]:
                        posterior *= self.likelihood[cls][col][x[col]]
                    else:
                        posterior *= 1e-6  # Laplace smoothing
                posteriors[cls] = posterior
            predictions.append(max(posteriors, key=posteriors.get))
        return predictions

    def predict_single(self, query):
        posteriors = {}
        for cls in self.classes:
            posterior = self.prior[cls]
            for col in range(len(query)):
                if query[col] in self.likelihood[cls][col]:
                    posterior *= self.likelihood[cls][col][query[col]]
                else:
                    posterior *= 1e-6  # Laplace smoothing
            posteriors[cls] = posterior

        # Convert probabilities to actual probabilities
        total_prob = sum(posteriors.values())
        probabilities = {cls: prob / total_prob for cls, prob in posteriors.items()}

        return max(probabilities, key=probabilities.get), probabilities


# Sample dataset
data = [
    ['Red', 'Casual', 'Domestic', 'Yes'],
    ['Red', 'Casual', 'Domestic', 'No'],
    ['Red', 'Casual', 'Domestic', 'Yes'],
    ['Yellow', 'Casual', 'Domestic', 'No'],
    ['Yellow', 'Casual', 'Imported', 'Yes'],
    ['Yellow', 'Formal', 'Imported', 'No'],
    ['Yellow', 'Formal', 'Imported', 'Yes'],
    ['Yellow', 'Formal', 'Domestic', 'No'],
    ['Red', 'Formal', 'Imported', 'No'],
    ['Red', 'Casual', 'Imported', 'Yes']
]

# Extract features and labels
X = [row[:-1] for row in data]
y = [row[-1] for row in data]

# Train the model
nb = NaiveBayesClassifier()
nb.fit(X, y)

# Define a specific query
query = ['Yellow', 'Casual', 'Domestic']

# Make a prediction for the specific query
predicted_class, probabilities = nb.predict_single(query)

print(f"Predicted class for the query: {predicted_class}")
print(f"Posterior probabilities: {probabilities}")
print(f"Sum of probabilities: {sum(probabilities.values())}")
```
